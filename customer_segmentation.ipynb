```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('customer.csv')

print df.head()
#stats
new_df=df.copy()
print new_df
print new_df.describe()
# filtering
print "filtering process starts "
print "printing all minimumm filterations nad maxi filtreations"
fresh_q3=16933.750000 
fresh_q1=3127.750000
frozen_q1=742.250000
frozen_q3=3554.250000
print new_df.loc[new_df.Fresh<fresh_q1].min()
# very low fresh and high gross
print new_df.loc[new_df.Frozen>frozen_q3].head(7)
#very high frozen and low detergent_paper
print new_df.loc[new_df.Frozen<frozen_q1].head()
#very high fresh and very very low frozen 
import numpy as np
import itertools
#removing outliers
outlier_list=[]

for i in new_df.columns:
    q1=np.percentile(new_df.loc[:,i],25)
    q3=np.percentile(new_df.loc[:,i],75)
    step=1.5*(q3-q1)
    print "Data points considered outliers for the feature '{}':".format(i)

    outliers_rows = new_df.loc[~((new_df[i] >= q1 - step) & (new_df[i] <= q3 + step)), :]
    outlier_list.append(list(outliers_rows.index))
    outliers = list(itertools.chain.from_iterable(outlier_list))

uniq_outlier=list(set(outliers))

#duplicate enteries removal

dup_outliers = list(set([x for x in outliers if outliers.count(x) > 1]))
print 'Outliers list:\n', uniq_outlier
print 'Length of outliers list:\n', len(uniq_outlier)

print 'Duplicate list:\n', dup_outliers
print 'Length of duplicates list:\n', len(dup_outliers)

real_data=new_df.drop(new_df.index[dup_outliers]).reset_index(drop=True)

print real_data.shape

# featres importances using DTR and RFR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
dep_vars=list(real_data.columns)

for i in dep_vars:
    
    new_data=real_data.drop([i],axis=1)
    new_features=pd.DataFrame(real_data.loc[:, i])
    x_train,x_test,y_train,y_test=train_test_split(new_data,new_features,test_size=.25,random_state=42)
    
    dectreg=DecisionTreeRegressor()
    dectreg.fit(x_train,y_train)
    score=dectreg.score(x_test,y_test)
    print " accuracy score for {}  as a dependent var {}".format(i,score)
    
# omitting grocery and detergent as we can predict true values of these things 60 to 70 % times 
# we are not omitting these values for some reason
# scaling the data with natural log before pca and kmeans 

log_data=np.log(real_data)

print "data after log scaling" 
sns.heatmap(log_data.corr(),annot=True,fmt=".2f") 
plt.show()

# we have seen that after correlation plotting there are only very few neg values
from sklearn.decomposition import PCA

pca=PCA()
pca.fit_transform(log_data)

print " PCA analysis"
print "variance"
var1=pca.explained_variance_
print var1
print "variance ratio" 
var2=pca.explained_variance_ratio_
print var2
principal_comp=pd.DataFrame(pca.fit_transform(log_data),index=log_data.index,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8'])
print principal_comp.head()
#transformation taling place
print "new data compressed one "
pca=PCA(n_components=2)
pca.fit_transform(log_data)
import matplotlib.pyplot as plt
print " PCA analysis"
print "variance"
var1=pca.explained_variance_
print var1
print "variance ratio" 
var2=pca.explained_variance_ratio_
print var2
principal_comp1=pd.DataFrame(pca.fit_transform(log_data),index=log_data.index,columns=['PC1','PC2'])
print principal_comp1.head()
dep=principal_comp1.iloc[:,0:2]
print dep

#confusion to use 3 pca compnents or  to use 2 , so we will use elbow method of  kmeans to get correction
wcss=[]
from sklearn.cluster import KMeans
x=range(2,11)
for i in x: 
 kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
 kmeans.fit(dep)
 wcss.append(kmeans.inertia_)
plt.plot(x,wcss,'-o',color='orange',label='elbow method')
plt.xlabel('no of clusters') 
plt.ylabel('sse (sum sq error) ')
plt.show()

# now using silhoette score we will analysis that whther our clustering will be better if we take 2 or 3 cluster 
from sklearn.metrics import silhouette_score
sc=[]
for i in x:
 kmeans=KMeans(n_clusters=i,random_state=0,init="k-means++")
 kmeans.fit(dep)
 pred=kmeans.predict(dep)
 score = silhouette_score(dep,pred,metric='euclidean')
 sc.append(score)
 print "For n_clusters = {}. The average silhouette_score is : {}".format(i, score)
plt.plot(x,sc,'-o',color="red",label="clustering analysis")
plt.xlabel('cluster')
plt.ylabel('silhouette score')
plt.show()

#Kmeans plotting channel and region are also there 
plt.Figure(figsize=(20,20))
kmeans=KMeans(n_clusters=2,random_state=0)
y=kmeans.fit_predict(dep)
print kmeans.labels_
centers=kmeans.cluster_centers_
principal_comp1["clusters"]=y

final_df=principal_comp1.copy()
print final_df
colormap = np.array(['red'])
colormap1 = np.array(['lime'])
plt.scatter(x=final_df[final_df.clusters==1]["PC1"], y=final_df[final_df.clusters==1]["PC2"], c=colormap,label='cluster1',alpha=0.54)
plt.scatter(x=final_df[final_df.clusters==0]["PC1"], y=final_df[final_df.clusters==0]["PC2"], c=colormap1,label='cluster2',alpha=0.54)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="yellow",label="centroid")
plt.title('K Mean Classification')
plt.ylim(-8,8)
plt.xlim(-8,8)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()

# Data recovery

#using inverse PCA-Transform

log_centers=pca.inverse_transform(centers)
true_centers=np.exp(log_centers)
segments = ['Segment {}'.format(i) for i in range(0,len(centers))]
true_centers = pd.DataFrame(np.round(true_centers), columns = real_data.columns)
true_centers.index = segments
print true_centers
#cluster deviation from median segment 0 has major spending on milk,grocery,detergent-paper restaurent and cafes
print true_centers-new_df.median()
#cluster deviation from mean segement 1 has major spending on fresh and frozen fresh food market and seafood or meat 
print true_centers-new_df.mean()

#predicting samples created at first stages 
print new_df.loc[new_df.Fresh<fresh_q1].min()
# very low fresh and high gross
print new_df.loc[new_df.Frozen>frozen_q3].head(7)
#very high frozen and low detergent_paper
print new_df.loc[new_df.Frozen<frozen_q1].head()

#samples 
indices = [43, 12, 39]

# Create a DataFrame of the chosen samples
# .reset_index(drop = True) resets the index from 0, 1 and 2 instead of 100, 200 and 300 
samples = pd.DataFrame(real_data.loc[indices], columns = real_data.columns).reset_index(drop = True)
print "Chosen samples of wholesale customers dataset:"
print samples

log_samp=np.log(samples)

pca=PCA(n_components=2)
pca_samp=pca.fit_transform(log_samp)
pca_data=pd.DataFrame(np.round(pca_samp, 4), columns = ['d1', 'd2'])
print pca_data
sample_preds = kmeans.predict(pca_samp)
for i, pred in enumerate(sample_preds):
    print "Sample point", i, "predicted to be in Cluster", pred
samples["clusters"]=sample_preds
pca_data["clusters"]=sample_preds

print samples
print pca_data
# clearly cluster 0 is for retailers like fresh market and meat food or sea food so 
# 1 sample is cluster 0 is correct and 1 also in cluster 0  as again retialers and 
# 2 sample is cluster 1 is correct like cafe / restaurent might be hotels

# final plotting 
plt.Figure(figsize=(20,20))
colormap = np.array(['red'])
colormap1 = np.array(['lime'])
plt.scatter(x=final_df[final_df.clusters==1]["PC1"], y=final_df[final_df.clusters==1]["PC2"], c=colormap,label='hotel/restaurent/cafe',alpha=0.24)
plt.scatter(x=final_df[final_df.clusters==0]["PC1"], y=final_df[final_df.clusters==0]["PC2"], c=colormap1,label='retailers/meat-food/sea-food markets',alpha=0.24)
plt.scatter(x=pca_data[pca_data.clusters==0]["d1"], y=pca_data[pca_data.clusters==0]["d2"],marker="x",label='0,1',alpha=0.80)
plt.scatter(x=pca_data[pca_data.clusters==1]["d1"], y=pca_data[pca_data.clusters==1]["d2"],marker="x",label='2',alpha=0.80)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="yellow",label="centroid")
plt.title('K Mean Classification')
plt.ylim(-8,8)
plt.xlim(-8,15)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()
```
